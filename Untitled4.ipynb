```
###############################################################################
###############################################################################
########  Getting   The   Necessary   Library  For  CNN   #####################
###############################################################################
###############################################################################
import numpy as np   
import matplotlib.pyplot as plt
import os
import cv2
# numpy for various array operations
# os to join directory paths and iterates through directories
# matplotlib just to show our images
# cv2 for image operations

###############################################################################
###############################################################################
###    Specifying     The     Data     Directory       ########################
###############################################################################
###############################################################################

DATADIR ="/content/BMWExamples"
CATEGORIES =["Class_1","Class_2","Class_3"]

for category in CATEGORIES:
    path = os.path.join(DATADIR,category) # This sends us to the class1,class2 or class3 directory
    for img in os.listdir(path): # Here we are getting the names of the bunch of images in the path directories
        img_array=cv2.imread(os.path.join(path,img)) #Now we have the full path joined together so that cv2 can read the images,it is converted to grayscale
        plt.imshow(img_array)  # Graphing our results to at least have an idea of what we are doing
        plt.show()
        break
    break
    
###############################################################################
###############################################################################
###  Taking    A  look  At   The   Img_Array      #############################
###############################################################################
###############################################################################

print(img_array) # We are observing that the images in img_array consist of a bunch of numbers
print(img_array.shape) # This provides the length and width of the images in question


###############################################################################
###############################################################################
### Reshaping Our Images For Images Processing  ###############################
###############################################################################
###############################################################################
IMG_SIZE = 100

new_array = cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
plt.imshow(new_array)  # Graphing our results to at least have an idea of what resize have done
plt.show()

###############################################################################
###############################################################################
####  Generating Our Training Dataset     #####################################
###############################################################################
###############################################################################

Training_data = []

def create_training_data():
    for category in CATEGORIES:
        path = os.path.join(DATADIR,category) # This sends us to the class1,class2 or class3 directory
        class_num = CATEGORIES.index(category)
        for img in os.listdir(path): # Here we are getting the names of the bunch of images in the path directories
            try:
                img_array=cv2.imread(os.path.join(path,img)) # Now we have the full path joined together so that cv2 can read the images,it is converted to grayscale
                new_array = cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
                Training_data.append([new_array,class_num])
            except Exception as e:
                pass
            
create_training_data()
           
###############################################################################
###############################################################################
####   Shuffle The Data So That Everything Becomes Random #####################
###############################################################################
###############################################################################

#import random
#random.shuffle(Training_data)

###############################################################################
###############################################################################
####  Checking Our Labels Associated With The Images   ########################
###############################################################################
###############################################################################

for sample in Training_data:
    print(sample[1])  # We printed this in order to show how good has the shuffle dealth with the categorical labels


###############################################################################
###############################################################################
####  Generating Your Train Feature And Train Labels Separately  ##############
###############################################################################
###############################################################################
    
X = [] # Train_features
y = [] # Train_labels

for features, label in Training_data:
    X.append(features)
    y.append(label)
    
X = np.array(X).reshape(-1, IMG_SIZE, IMG_SIZE, 3)

###############################################################################
###############################################################################
###   SAVING    OUR      DATASET    ###########################################
###############################################################################
###############################################################################
import pickle

pickle_out = open("X.pickle","wb")
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open("y.pickle","wb")
pickle.dump(y, pickle_out)
pickle_out.close()

###############################################################################
###############################################################################
## Performing One Hot Encoding  ###############################################
###############################################################################
###############################################################################
from keras.utils import to_categorical
y = to_categorical(y)



###############################################################################
###############################################################################
### DEALING WITH THE CONVOLUTIONAL NEURAL NETWORK MODELS ######################
###############################################################################
###############################################################################
import keras
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten,MaxPooling2D,Dropout
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
import pickle
from tensorflow.keras import optimizers

pickle_in = open("X.pickle","rb")
X = pickle.load(pickle_in)

pickle_in = open("y.pickle","rb")
y = pickle.load(pickle_in)



###############################################################################
###############################################################################
###    Normalization     Of    Our       Dataset     ##########################
###############################################################################
###############################################################################


X = X/255.0

from keras.utils import to_categorical
y = np_utils.to_categorical(y,3)


model  =  Sequential()
model.add( Conv2D(32, kernel_size=3, activation='relu', input_shape= (100,100,3)) )
model.add(Conv2D(32, kernel_size=3, activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(rate=0.25))
model.add(Conv2D(64, kernel_size=3, activation='relu'))
model.add(Conv2D(64, kernel_size=3, activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(rate=0.25))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(rate=0.25))
model.add(Dense(256, activation='relu'))
model.add(Dropout(rate=0.25))
model.add(Dense(3, activation='softmax'))





model.compile(loss="categorical_crossentropy",
              optimizer="adam",
              metrics =['acc'])

###############################################################################
###############################################################################
###     Saving     The    Model    In    TFlite   Format    ###################
###############################################################################
###############################################################################





###############################################################################
###############################################################################
### Performing  The Fit Associated With The Model ############################
###############################################################################
###############################################################################

History = model.fit(
    X,
    y,
    batch_size=5,
    epochs=200,
    validation_split=0.1)


###############################################################################
###############################################################################
##  Performing    A     Plot    Associated    With    The   Loss        #######
###############################################################################
###############################################################################

plt.plot(History.history['loss'])
plt.plot(History.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['loss', 'val_loss'], loc='upper left')
plt.show()


###############################################################################
###############################################################################
###     Performing  A  Plot  Associated  With  The   Accuracy       ###########
###############################################################################
###############################################################################

# Accuracy
plt.plot(History.history['acc'])
plt.plot(History.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['acc', 'val_acc'], loc='upper left')
plt.show()


###############################################################################
###############################################################################
####      Saving       The        Model       #################################
###############################################################################
###############################################################################

model.save("/content/BMWExamples/BMW_kof.h5")



###############################################################################
###############################################################################
##     Saving      The      Model     As     Tflite         ###################
###############################################################################
###############################################################################
#
import tensorflow as tf
converter = tf.lite.TFLiteConverter.from_keras_model_file( '/content/BMWExamples/BMW_kof.h5' ) # Your model's name
model = converter.convert()
file = open( '/content/BMWExamples/BMW.tflite' , 'wb' ) 
file.write(model)



###############################################################################
###############################################################################
##### Performing Prediction Using Our Model  ##################################
###############################################################################
###############################################################################

import cv2
import tensorflow as tf

CATEGORIES =["Class_1","Class_2","Class_3"]


def prepare(filepath):
    IMG_SIZE=100
    Img_array=cv2.imread(filepath)
    New_array=cv2.resize(Img_array,(IMG_SIZE,IMG_SIZE))
    return New_array.reshape(-1,IMG_SIZE,IMG_SIZE,3)


###############################################################################
###############################################################################
#######     Loading        The       Models      ##############################
###############################################################################
###############################################################################

Model = tf.keras.models.load_model("/content/BMWExamples/BMW_kof.h5")


###############################################################################
###############################################################################
######  Performing The First Prediction #######################################
###############################################################################
###############################################################################

Prediction=Model.predict([prepare("/content/BMWExamples/Test/2.jpg")])
print(CATEGORIES[np.argmax(Prediction)])



    









    
```
